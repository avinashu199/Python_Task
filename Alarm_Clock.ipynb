```python
'''
Alaram clock program
1.current time
2.alaram time
3.message when it reached alaram time
'''
```

```python
import datetime
import time
def current_time(at):
    while True:
        ct=datetime.datetime.now().strftime("%H:%M:%S")
        time.sleep(1)
        print(ct)
        if ct==at:
            print("Wake up !!!")
            break
def alaram_time():
    at=input("Enter the alarm time[HH:MM:SS]")
    print(f"You have set alaram at {at}")
    return at
at=alaram_time()
current_time(at)
```
