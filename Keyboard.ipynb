```python
#Mobie keyboard-- we have fixed values at each places and which cannot be changed/altered. so we can use the tuple
row1=(1,2,3)
row2=(4,5,6)
row3=(7,8,9)
row4=('*',0,'#')
keyboard=((row1),
          (row2),
          (row3),
          (row4))
for i in keyboard:
    for j in i:
        print(j,end='')
    print()
    
```

```python
#banking key board - where numbers are randomly changes , we can acheive this by using set where order is changed randomly
row1={1,2,3}
row2={4,5,6}
row3={7,8,9}
row4={'*',0,'#'}
keyboard=[{row1},
          {row2},
          {row3},
          {row4}]
for i in keyboard:
    for j in i:
        print(j,end='')
    print()
```
