```python
#Make shopping cart program
'''
item details
price
quantity
total price
'''
item=[]
price=[]
quantity=[]
total=[]
while True:
    user=input("Enter the details of the item[Press q to quit]:")
    if user.lower()=='q':
        break
    else:
        item.append(user)
while True:
    a=float(input("Enter the price of the item:[Enter negative numbers to quit]:"))
    if a<=0:
        break
    else:
        price.append(a)
while True:
    b=int(input("Enter the quantity:[Press q to quit]:"))
    if b<0:
        break
    else:
        quantity.append(b)
print(item)
print(price)
print(quantity)
for i in range(len(price)):
            c=price[i]*quantity[j]
            total.append(c)
print(total)
print(f"The total items you have ordered are {item}")
print(f"The total cart value is{sum(total)}")
```
