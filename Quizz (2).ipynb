```python
'''
Quizz program -- final output should also give me the score
'''
questions=["1. Who is known as the God of Cricket?","2. Which country won the ICC Cricket World Cup 2019?","3. How many runs are awarded for hitting the ball over the boundary without bouncing?","4. Who holds the record for the highest individual score in ODI cricket?","5. What is the maximum number of overs allowed per bowler in a 50-over match?"]
options=["A. Ricky Ponting | B. Brian Lara | C. Sachin Tendulkar | D. Virat Kohli","A. England | B. New Zealand | C. Australia | D. India","A. 4 | B. 5 | C. 6 | D. 3","A. Chris Gayle | B. Martin Guptill | C. Rohit Sharma | D. David Warner","A. 15 | B. 10 | C. 12 | D. 8"]
correctans=["C","A","C","C","B"]
userans=[]
option=["A","B","C","D"]
count=0
for i in range(len(questions)):
            print(questions[i])
            print(options[i])
            ans=input("Enter the option[A/B/C/D]:")
            while ans not in option:
                ans=input("Enter the option only from suggestions[A/B/C/D]:") 
            userans.append(ans)
for x in range(len(userans)):
    if userans[x]==correctans[x]:
        count=count+20
print(f"You have scored:{count}%")
```
