User name should be no more than 12 characters
user name should not contain spaces
user name should not contain digits


```python
name=input("Enter the user name:")
#user name should be less than 12 characters
if len(name)<12:
    if name.isalpha() and not name.isdigit():
        print("Valid user name")
    else:
        print("User name contains either space or digits. please re enter without space and digits")
else:
    print("User name has characters more than 12. please re enter again")
        
```
