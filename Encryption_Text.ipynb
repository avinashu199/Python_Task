```python
import string  #String module consists of the constants and functions which are by defualt available in the python
import random
text=" "+string.ascii_lowercase+string.digits


#ascii_lowercase-- this constant have all the values of individual letters
#digits-- this constant is having all the digits to work with

#convert text from string to list using typecasting
text_list=list(text)
#to create the encrypted key use the random method
key=text_list.copy()

random.shuffle(key)
print(f"Original list:{text_list}")
print(f"Updated keyst:{key}")
while True:
    encrypted=''
    message=input("Enter the message:")
    action=input("Enter the action[encrypt/decrypt]:")
    if action=='encrypt':
        for i in message:
            #get the index position from the original list
            index=text_list.index(i)
            encrypted=encrypted+key[index]
        print(encrypted)
    elif action=='decrypt':
        for i in message:
            #get the index position from the original list
            index=key.index(i)
            encrypted=encrypted+text_list[index]
        print(encrypted)
    else:
        break
        
```
