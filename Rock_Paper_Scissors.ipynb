```python
import random
elements=('Rock','Paper','Scissors')
Computer=random.choice(elements)
while True:
    Player=input("Start the game by entering the choice[Rock/Paper/Scissors]:")
    if Player==Computer:
        print("It's a Tie!!!")
        break
    elif Player=='Rock' and Computer=='Paper':
        print("You loose!!!")
        break
    elif Player=='Rock' and Computer=='Scissors':
        print("You win!!!")
        break
    elif Player=='Paper' and Computer=='Rock':
        print("You win!!!")
        break
    elif Player=='Paper' and Computer=='Scissors':
        print("You loose!!!")
        break
    elif Player=='Scissors' and Computer=='Rock':
        print("You loose!!!")
        break
    elif Player=='Scissors' and Computer=='Paper':
        print("You win!!!")
        break
    else:
        print("Entered an invalid elements of the game")
        break
print(f"Player:{Player} vs Computer:{Computer}")
```
