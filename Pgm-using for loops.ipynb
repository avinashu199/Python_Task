```python
'''
*
**
***
****
'''
lines=int(input("Enter the number of lines you need"))
for i in range(lines):
    for j in range(lines):
        if j<=i:
            print('*',end=' ')
    print('')
```

```python
#combination of both programs
'''
*
**
***
****
***
**
*
'''
lines=int(input("Enter the number of lines you need"))
for i in range(lines):
    for j in range(lines):
        if j<=i:
            print('*',end=' ')
    print('')
for i in range(lines-1):
    for j in range(lines-1):
        if j>= i:
            print('*',end=' ')
    print('')
```

```python
'''
  *
 **
*** 
'''
lines=int(input('Enter the maximum number of lines:'))
for i in range(lines):
    for j in range(lines):
        if i+j<lines-1:
            print(' ',end='')
        else:
            print('*',end='')
    print('')
        
```
