```python
import time
import datetime
directions=['Gandhinagar','Singhnagar','Elur Road','Bandar Road']
def yellow(road):
    print(f"{road}:游리 - WAIT")
    counter=0
    while counter<3:
        c_dt=datetime.datetime.now()
        c_dt=c_dt.strftime("%H:%M:%S")
        print(c_dt)
        time.sleep(1)
        counter+=1  
def stop(road):
    print(f"{road}:游댮 - STOP ")

def start(road):
    print(f"{road}: 游릭  - GO")
    counter=0 
    while counter<5:
        c_dt=datetime.datetime.now()
        c_dt=c_dt.strftime("%H:%M:%S")
        print(c_dt)
        time.sleep(1)
        counter+=1
    yellow(road)
for direction in directions:
    start(direction)
    stop(direction)
```
