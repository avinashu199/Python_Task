```python
'''
Concession stand program
'''
menu={"Popcorn":6,"Hot dog":2,"Bottled Water":1,"Candy":1}
items=menu.keys()
cart=[]
value=0
print("--------MENU---------")
for key,value in menu.items():
    print(f"{key:<15}:${value:.2f}")
print("---------------------")
while True:
    user=input("Enter the item[Select q to quit]:")
    if user=='q':
        break
    elif user in items:
       cart.append(user)
    else:
        print("Enter the re enter the items correctly from the menu:")
print(cart)
for total in cart:
    value=value+menu[total]
print(value)
```
