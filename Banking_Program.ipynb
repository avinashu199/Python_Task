```python
balance=0
def show_balance():
    print(f"Dear Account holder! your available balance is${balance}")
def deposit():
    global balance
    amount=int(input("Enter the amount to deposit:"))
    if amount>0:
        balance=balance+amount
        print(f"Thank you! You have successfully deposited :${amount}")
    else:
        print("Please retry again with valid deposit amount")
def withdraw():
    global balance
    withdrawl=int(input("Enter the amount for withdrawl"))
    if withdrawl>0 and withdrawl<=balance:
        balance=balance-withdrawl
        print(f"Thank you! You have successfully withdrawl:${withdrawl}")
    else:
        print("Insufficient funds!! Please retry again")
def exit():
    pass
while True:
    print("************************")
    print("1.Show balance")
    print("2.Depost")
    print("3.Withdraw")
    print("4.Exit")
    print("************************")
    choice=int(input("Select valid option[1-4]"))
    match choice:
        case 1:
            show_balance()
        case 2:
            deposit()
        case 3:
            withdraw()
        case 4:
            exit()
            break
    
```
