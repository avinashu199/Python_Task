```python
'''
Number guessing
'''
import random
actnum=random.randint(1,100)
while True:
    expnum=int(input("Guess the number[Between 1 & 100]:"))
    if actnum==expnum:
        print("Hurray !! you have guessed the number correctly")
        break
    elif actnum>expnum:
        print("Please guess a higher number again")
    elif actnum<expnum:
        print("Please guess a lower number again")
print(f"The number that is gueesed correctly is {actnum}")
```
